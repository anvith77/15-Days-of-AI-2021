```python
import numpy as np
import matplotlib.pyplot as plt  # To visualize
import pandas as pd  # To read data
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('/kaggle/input/szeged-weather/weatherHistory.csv')
data = data.rename(columns={'Temperature (C)': 'Temperature'})
X = data.Temperature.values.reshape(-1, 1)  # values converts it into a numpy array
Y = data.Humidity.values.reshape(-1, 1)  # -1 means that calculate the dimension of rows, but have 1 column
linear_regressor = LinearRegression()  # create object for the class
linear_regressor.fit(X, Y)  # perform linear regression
Y_pred = linear_regressor.predict(X)  # make predictions
```

```python
plt.scatter(X, Y)
plt.plot(X, Y_pred, color='red')
plt.show()
```
